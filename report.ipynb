```python
import pandas as pd

df = pd.read_csv('./cleaned_data.csv')
df.head(5)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Assuming df is your DataFrame

# Create a subplot with 2 rows and 2 columns
fig = make_subplots(rows=2, cols=2, subplot_titles=['Boost Clock', 'Max Resolution', 'Width', 'Height'])

# Add boxplot for 'Boost_clock' in the first subplot
fig.add_trace(go.Box(y=df['Boost_clock'], name='Boost Clock'), row=1, col=1)

# Add boxplot for 'Max_resolution' in the second subplot
fig.add_trace(go.Box(y=df['Max_resolution'], name='Max Resolution'), row=1, col=2)

# Add boxplot for 'Width' in the third subplot
fig.add_trace(go.Box(y=df['Width'], name='Width'), row=2, col=1)

# Add boxplot for 'Height' in the fourth subplot
fig.add_trace(go.Box(y=df['Height'], name='Height'), row=2, col=2)

# Update layout to prevent overlapping
fig.update_layout(height=600, width=1000)

# Initialize Dash app
app = dash.Dash(__name__)

# Define layout of the dashboard
app.layout = html.Div([
    html.H1("Your Dashboard Title"),
    
    # Add the Plotly figure to the layout
    dcc.Graph(figure=fig),
    
    # Add any other components or controls as needed
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
