```python
import pandas as pd

df = pd.read_csv('./rawdata.csv')
df.head()
```

```python
columns = ['STT', 'Name', 'Manufacturer', 'Brand', 'Condition', 'Price', 'Warranty_type', 'Warranty_time', 'Architecture', 'Max_resolution', 'Base_clock', 'Boost_clock', 'Memory_clock', 'Memory_size', 'Memory_type', 'Bus_width', 'Bandwidth', 'Cores', 'Length', 'Width', 'Height', 'Bus_interface', 'TDP', 'OpenGL', 'DirectX']
df = df.set_axis(columns, axis=1).drop(columns=['STT'])
df
```

```python
import numpy as np

df.replace(['?', '.', '_', '-', '', 'None', 'NULL', 'Not', '|'], np.nan, inplace=True)
df.isnull().sum()
```

```python
missing_values = df[df.isnull().any(axis=1)]
missing_values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Plot boxplot for 'Boost_clock' in the first subplot
sns.boxplot(df['Boost_clock'], ax=axes[0, 0])
axes[0, 0].set_title('Boost Clock')

# Plot boxplot for 'Max_resolution' in the second subplot
sns.boxplot(df['Max_resolution'], ax=axes[0, 1])
axes[0, 1].set_title('Max Resolution')

# Plot boxplot for 'Width' in the third subplot
sns.boxplot(df['Width'], ax=axes[1, 0])
axes[1, 0].set_title('Width')

# Plot boxplot for 'Height' in the fourth subplot
sns.boxplot(df['Height'], ax=axes[1, 1])
axes[1, 1].set_title('Height')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

Ta thấy giá trị của Boost Clock, Width, Height phân bố đều, có tính đối xứng, cho nên ta sử dụng mean để điền khuyết. Giá trị của Max Resolution bị skewed và là giá trị phân loại nên ta điền khuyết bằng mode

```python
width_mean = df["Width"].astype("float").mean()
df["Width"].replace(np.nan, width_mean, inplace=True)

height_mean = df["Height"].astype("float").mean()
df["Height"].replace(np.nan, height_mean, inplace=True)

BoostClock_mean = df["Boost_clock"].astype("float").mean()
df["Boost_clock"].replace(np.nan, BoostClock_mean, inplace=True)
```

```python
df["Max_resolution"].replace(np.nan, df["Max_resolution"].value_counts().idxmax(), inplace=True)
```

```python
df.info()
```

```python
df1 = pd.get_dummies(df["Condition"])
df['IsNew'] = df1['Mới']
df.drop('Condition', axis=1, inplace=True)
```

```python
df2 = pd.get_dummies(df["Warranty_type"])
df['No_warranty'] = df2['Không có bảo hành']
df['Genuine_warranty'] = df2['Bảo hành hãng']
df['Seller_warranty'] = df2['Người bán bảo hành']
df.drop('Warranty_type', axis=1, inplace=True)
```

```python
df.to_csv('./cleaned_data.csv',index=False)
```
